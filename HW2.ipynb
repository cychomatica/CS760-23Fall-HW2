### Q2.7

```python
raw_data_big = np.loadtxt('Dbig.txt')
np.random.seed(2023)
rand_idx = np.random.choice(np.arange(raw_data_big.shape[0]), size=raw_data_big.shape[0], replace=False)
D_32 = raw_data_big[rand_idx[:32]]
D_128 = raw_data_big[rand_idx[:128]]
D_512 = raw_data_big[rand_idx[:512]]
D_2048 = raw_data_big[rand_idx[:2048]]
D_8192 = raw_data_big[rand_idx[:8192]]
D_test = raw_data_big[rand_idx[8192:]]

DTs, errs = [], []
for D_n, n in zip([D_32, D_128, D_512, D_2048, D_8192], [32, 128, 512, 2048, 8192]):
    DT = decision_tree.MakeSubtree(D_n)
    DTs.append(DT)
    err = 0.
    for example in D_test:
        x, y = example[:-1], example[-1]
        y_pred = DT.classifiy(x)
        err += int(y_pred != y) / D_test.shape[0]
    err = err * 100
    errs.append(err)
    print('n: {}\tnumber of nodes: {}\terror rate: {:4f}%'.format(n, DT.len(), err))

import matplotlib.pyplot as plt
plt.figure()
plt.plot([32,128,512,2048,8192], errs)
plt.xlabel('n')
plt.ylabel('error rate')
plt.savefig('n-err-curve.pdf')
plt.show()
plt.close()

for DT_n, n in zip(DTs, [32, 128, 512, 2048, 8192]):
    visualization.draw_decision_boundary(model_function=DT_n.classifiy,
                                         x1_grid_min=raw_data_big[:,0].min(),
                                         x1_grid_max=raw_data_big[:,0].max(),
                                         x2_grid_min=raw_data_big[:,1].min(),
                                         x2_grid_max=raw_data_big[:,1].max(),
                                         title='Model Decision Boundary trained on D_{}'.format(n), savefile='D_{}_decision.pdf'.format(n))
```

### Q3

```python
from sklearn import tree

raw_data_big = np.loadtxt('Dbig.txt')
np.random.seed(2023)
rand_idx = np.random.choice(np.arange(raw_data_big.shape[0]), size=raw_data_big.shape[0], replace=False)
D_32 = raw_data_big[rand_idx[:32]]
D_128 = raw_data_big[rand_idx[:128]]
D_512 = raw_data_big[rand_idx[:512]]
D_2048 = raw_data_big[rand_idx[:2048]]
D_8192 = raw_data_big[rand_idx[:8192]]
D_test = raw_data_big[rand_idx[8192:]]

sklearn_DTs, sklearn_errs = [], []
for D_n, n in zip([D_32, D_128, D_512, D_2048, D_8192], [32, 128, 512, 2048, 8192]):
    sklearn_DT = tree.DecisionTreeClassifier()
    sklearn_DT.fit(D_n[:,:-1], D_n[:, -1])

    sklearn_DTs.append(sklearn_DT)
    err = 0.
    x_test, y_test = D_test[:, :-1], D_test[:, -1]
    y_pred = sklearn_DT.predict(x_test)
    err += sum(y_pred != y_test) / D_test.shape[0]
    err = err * 100
    sklearn_errs.append(err)
    print('n: {}\tnumber of nodes: {}\terror rate: {:4f}%'.format(n, sklearn_DT.tree_.node_count, err))

import matplotlib.pyplot as plt
plt.figure()
plt.plot([32,128,512,2048,8192], sklearn_errs)
plt.xlabel('n')
plt.ylabel('error rate')
plt.savefig('sklearn-n-err-curve.pdf')
plt.show()
plt.close()
```

### Q4

```python
from scipy.interpolate import lagrange
from numpy.polynomial.polynomial import Polynomial
import matplotlib.pyplot as plt
import math
import numpy as np

a, b, n = 0, math.pi, 100

np.random.seed(2023)
x_uniform_train = np.random.uniform(a, b, 100)
y_uniform_train = np.sin(x_uniform_train)

np.random.seed(2023)
x_uniform_test = np.random.uniform(a, b, n)
y_uniform_test = np.sin(x_uniform_test)

poly = lagrange(x_uniform_train, y_uniform_train)
y_uniform_train_pred = poly(x_uniform_train)
y_uniform_test_pred = poly(x_uniform_test)
sin_train_err = ((y_uniform_train_pred - y_uniform_train)**2).mean()
sin_test_err = ((y_uniform_test_pred - y_uniform_test)**2).mean()
print('train MSE: {}'.format(sin_train_err))
print('test MSE: {}'.format(sin_test_err))

# std = 1
std_max = 50
errs = []
for std in range(1, std_max):
    noise = np.random.randn(n)
    x_train_noisy = x_uniform_train + std * noise
    y_train_noisy = np.sin(x_train_noisy)
    poly = lagrange(x_train_noisy, y_train_noisy)
    y_uniform_pred = poly(x_uniform_test)
    sin_err = ((y_uniform_pred - y_uniform_test)**2).mean()
    errs.append(sin_err)
    print('test MSE with noise std={}: {}'.format(std, sin_err))
plt.figure()
plt.plot(np.arange(1,std_max), [math.log(err, 10) for err in errs])
plt.xlabel('std')
plt.ylabel('log$_{10}$MSE')
plt.savefig('std-err-curve.pdf')
plt.show()
plt.close()
# sin_err
# x_uniform_train
# y_uniform_train
```
